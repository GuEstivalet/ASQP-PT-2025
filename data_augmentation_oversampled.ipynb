```python
from google.colab import drive
drive.mount("/content/drive")

import ast
import re
import random
import numpy as np
import pandas as pd
from pathlib import Path
```

```python
SEED = 42
random.seed(SEED)
np.random.seed(SEED)

DATA_PATH = Path("/content/drive/MyDrive/ASQP/data_processado/train_processed.csv")
OUT_DIR = Path("/content/drive/MyDrive/ASQP/data")

OUT_OVERS = OUT_DIR / "treino_overs.csv"

TEXT_COL = "text"
INPUT_COL = "input"
TARGET_COL = "target"
QUAD_COL = "quadruples"

AUGMENTED_COL = "augmented"
SOURCE_ID_COL = "source_id"

TARGET_MULTIPLIER = 3
```

```python
df = pd.read_csv(DATA_PATH)

df.columns = [c.strip() for c in df.columns]

required_cols = ["id",TEXT_COL,INPUT_COL,TARGET_COL,QUAD_COL]

missing = [c for c in required_cols if c not in df.columns]

if missing:
    raise ValueError(f"Colunas ausentes: {missing}")

df = df.reset_index(drop=True)

BASE_SIZE = len(df)
TARGET_SIZE = BASE_SIZE * TARGET_MULTIPLIER

print("Base:", BASE_SIZE)
print("Alvo:", TARGET_SIZE)
display(df.head())
```

```python
# replica 3x
overs_df = pd.concat(
    [df.copy(), df.copy(), df.copy()],
    ignore_index=True
)

# atualiza ids
overs_df["source_id"] = overs_df["id"]
overs_df["id"] = [
    f"{sid}_overs_{i // len(df)}"
    for i, sid in enumerate(overs_df["source_id"])
]

# embaralha
overs_df = overs_df.sample(
    frac=1,
    random_state=42
).reset_index(drop=True)

overs_df.to_csv(
    OUT_OVERS,
    index=False,
    encoding="utf-8"
)

print("Original:", df.shape)
print("Oversampling:", overs_df.shape)
print("Salvo em:", OUT_OVERS)
```
